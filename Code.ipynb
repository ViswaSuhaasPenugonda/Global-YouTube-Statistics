```python
import pandas as pd
```

```python
# Read in the CSV data using pandas and display the first 10 rows
df = pd.read_csv('Data.csv')
df.head(10)
```

```python
# Standardize column names - convert to lowercase, remove spaces, etc
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
# Check for any missing values
df.isnull().sum()
```

```python
# List of columns to drop
columns_to_drop = ['category','title','abbreviation','video_views_rank', 'created_year', 'created_month', 'created_date',
                   'gross_tertiary_education_enrollment_(%)', 'urban_population', 'latitude', 'longitude']

# Drop the specified columns
df = df.drop(columns=columns_to_drop)
```

```python
# Deal with the missing values
df['country'].fillna(df['country'].mode()[0], inplace=True)
df['channel_type'].fillna(df['channel_type'].mode()[0], inplace=True)
df['country_rank'].fillna(df['country_rank'].mean(), inplace=True)
df['channel_type_rank'].fillna(df['channel_type_rank'].mean(), inplace=True)
df['video_views_for_the_last_30_days'].fillna(df['video_views_for_the_last_30_days'].mean(), inplace=True)
df['subscribers_for_last_30_days'].fillna(df['subscribers_for_last_30_days'].mean(), inplace=True)
df['population'].fillna(df['population'].mean(), inplace=True)
df['unemployment_rate'].fillna(df['unemployment_rate'].mean(), inplace=True)

df.isnull().sum()
```

```python
# Handle duplicate rows if any
df.drop_duplicates(inplace=True)
```

```python
# Understand the data type of each column
df.info()
```

```python
# Convert float columns to integer
int_cols = ['video_views', 'country_rank', 'channel_type_rank', 'video_views_for_the_last_30_days',
            'subscribers_for_last_30_days', 'population']

for col in int_cols:
    df[col] = df[col].astype(int)

df.info()
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv("Data_cleaned.csv", index=False)
```
